```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ResNet_0.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
df['train loss'].plot(ax=ax)
df['test loss'].plot(ax=ax)
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
df['train acc'].plot(ax=ax)
df['test acc'].plot(ax=ax)
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```
